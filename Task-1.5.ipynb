```python
# take dynamic input and reverse without using built-in reversal functions
user_str = input("Enter a string: ")
reversed_str = ""
i = len(user_str) - 1
while i >= 0:
    reversed_str += user_str[i]
    i -= 1  # move to previous character
print(reversed_str)
```

```python
# print the next prime number after a given integer (no functions)

n = int(input("Enter an integer: "))
candidate = n + 1

while True:
    if candidate < 2:
        candidate += 1
        continue

    is_prime = True

    if candidate % 2 == 0 and candidate != 2:
        is_prime = False
    else:
        limit = int(candidate ** 0.5) + 1
        p = 3
        while p <= limit:
            if candidate % p == 0:
                is_prime = False
                break
            p += 2

    if is_prime:
        print(candidate)
        break

    candidate += 1
```
